```python
# Imports of Fun

# Numpy and PyPlot
import numpy as np
import matplotlib.pyplot as plt

# Inline outputs of matplotlib
%matplotlib inline
```

```python
def velocity_potential(x, y, Λ, x_s, y_s):
    distances = np.sqrt((x - x_s[:, np.newaxis, np.newaxis])**2 + (y - y_s[:, np.newaxis, np.newaxis])**2)
    return np.sum(Λ[:, np.newaxis, np.newaxis] / (2 * np.pi) * np.log(distances), axis=0)
```

```python
def stream_function(x, y, Λ, x_s, y_s):
    φ = velocity_potential(x, y, Λ, x_s, y_s)
    ψ_x = np.gradient(φ, axis=0)
    ψ_y = np.gradient(φ, axis=1)
    return ψ_y[::-1, :]
```

```python
Λ = np.array([-1, -3, 1, 2, 1]) * np.pi
x_s = np.array([2, 1, 0, -1, -2])
y_s = np.zeros_like(x_s)

x = np.linspace(-4, 4, 101)
y = np.linspace(-4, 4, 101)
X, Y = np.meshgrid(x, y)

ψ = stream_function(X, Y, Λ, x_s, y_s)
```

```python
plt.contour(X, Y, ψ, levels=np.linspace(-5, 5, 101), cmap='viridis')
plt.xlim(-2.5, 2.5)
plt.ylim(-2.5, 2.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
