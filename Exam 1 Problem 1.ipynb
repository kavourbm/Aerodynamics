```python
V = 1
R = .5
x = np.linspace(-10, 10, 101)
y = np.linspace(10, 10, 101)
X, Y = np.meshgrid(x, y)
lamb = np.zeros([101,101])
count1 = 0
count2 = 0
for i in x:
    for j in y:
        lamb[count1,count2] = (V*np.sqrt(i**2+j**2)*(np.sin(np.arctan(j/i)*(180/np.pi)))*(1-(R**2/(i**2+j**2)))+2/(2*np.pi)*np.log(np.sqrt(i**2+j**2)/R))
        print(np.sqrt(i**2+j**2),np.arctan(j/i)*(180/np.pi),i,j)
        print(lamb[count1,count2])
        count2 += 1
    count1 += 1
    count2 = 0
```

```python
plt.contour(X,Y,lamb,cmap="RdGy")
plt.colorbar()
plt.show()
```
